# Logistic Regression

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Load Data

stu_data = pd.read_csv('combined_csv.csv')
stu_data.head()
```

```python
# Transform data to fit model

#stu_data = stu_data.drop(columns=['parental level of education'])

# def parse_values(x):
#     if x > 90:
#        return "A"
#     elif x > 80:
#        return "B"
#     elif x > 70:
#        return "C"
#     elif x > 64:
#         return "D"
#     else:
#         return "F"

# # Generate binary pass/fail values or Generate a letter grade based on actual scores.

# stu_data['math grade'] = stu_data['math score'].apply(parse_values)
# stu_data['reading grade'] = stu_data['reading score'].apply(parse_values)
# stu_data['writing grade'] = stu_data['writing score'].apply(parse_values)
stu_data['pass math'] = stu_data['math score'].apply(lambda x: 1 if x>64 else 0)
stu_data['pass reading'] = stu_data['reading score'].apply(lambda x: 1 if x>64 else 0)
stu_data['pass writing'] = stu_data['reading score'].apply(lambda x: 1 if x>64 else 0)
stu_data
```

```python
# One hot encode dummy values
stu_data = pd.get_dummies(stu_data)
# Drop unnecessary values
stu_data = stu_data.drop(columns=['gender_male', 'lunch_standard', 'test preparation course_none'])
stu_data = stu_data.drop(columns=['math score', 'reading score', 'writing score'])
stu_data
```

```python
# Set target value


X = stu_data.drop('pass math', axis=1)
y = stu_data['pass math']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:25]}")
print(f"First 10 Actual labels: {y_test[:25].tolist()}")
```

```python
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head(25)
```
