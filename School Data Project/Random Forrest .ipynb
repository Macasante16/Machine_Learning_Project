```python
# Read the CSV file into pandas DF

stu_data = pd.read_csv('combined_csv.csv')
stu_data.head(20)
```

```python
stu_data = stu_data.drop(columns=['race/ethnicity', 'parental level of education'])
stu_data
```

```python
stu_data = pd.get_dummies(stu_data)
stu_data.head()
```

```python
stu_data = stu_data.drop(columns=['gender_male', 'lunch_standard', 'test preparation course_none'])
stu_data.head()
```

```python
# # Pass = 1
# # Fail = 0

stu_data['pass_math'] = stu_data['math score'].apply(lambda x: 1 if x>64 else 0)
stu_data['pass_reading'] = stu_data['reading score'].apply(lambda x: 1 if x>64 else 0)
stu_data['pass_writing'] = stu_data['writing score'].apply(lambda x: 1 if x>64 else 0)
```

```python
stu_data.head()
```

```python
stu_data = stu_data.drop(columns=['math score', 'reading score', 'writing score'])
stu_data
```

```python
target = stu_data['pass_writing']
target_names = ['fail', 'pass']
```

```python
data = stu_data.drop('pass_writing', axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=5000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
